

### 1. What is a random variable in probability theory?

A random variable is a numerical value associated with the outcome of a random experiment. It maps outcomes of a probabilistic scenario to numbers. For example, when rolling a die, the result (1 to 6) can be seen as a random variable.

---

### 2. What are the types of random variables?

There are two types:
- **Discrete Random Variables**: Take specific, countable values (e.g., number of heads in coin tosses).
- **Continuous Random Variables**: Take an infinite number of possible values within a range (e.g., height, weight).

---

### 3. What is the difference between discrete and continuous distributions?

- **Discrete Distribution**: Describes the probability of outcomes of a discrete random variable.
- **Continuous Distribution**: Represents probabilities as areas under a curve for continuous variables.

---

### 4. What are probability distribution functions (PDF)?

A **Probability Distribution Function (PDF)** describes the likelihood of a random variable taking specific values.
- For **discrete variables**, it's called **PMF (Probability Mass Function)**.
- For **continuous variables**, PDF is a function whose area under the curve between two values represents probability.

---

### 5. How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?

- **PDF** shows the probability density at a specific value.
- **CDF** gives the probability that a variable is less than or equal to a value.
  \[
  CDF(x) = P(X \leq x)
  \]

---

### 6. What is a discrete uniform distribution?

A discrete uniform distribution assigns equal probability to each outcome in a finite set. For example, in a fair die roll, each face (1 to 6) has equal probability \( \frac{1}{6} \).

---

### 7. What are the key properties of a Bernoulli distribution?

- Two possible outcomes: success (1) or failure (0).
- Single trial.
- Probability of success is \( p \), and failure is \( 1-p \).
- Mean = \( p \), Variance = \( p(1-p) \)

---

### 8. What is the binomial distribution, and how is it used in probability?

Binomial distribution describes the number of successes in a fixed number of independent Bernoulli trials. It’s defined as:
\[
P(X = k) = \binom{n}{k} p^k (1-p)^{n-k}
\]
Used in modeling binary outcomes, like pass/fail scenarios.

---

### 9. What is the Poisson distribution and where is it applied?

Poisson distribution models the number of events happening in a fixed interval of time or space, given the events occur independently and at a constant average rate \( \lambda \).
Used in:
- Call arrivals at a call center.
- Decay events per minute.

---

### 10. What is a continuous uniform distribution?

It’s a distribution where all values in a given range are equally likely.
For a range [a, b], the PDF is:
\[
f(x) = \frac{1}{b-a} \text{ for } a \le x \le b
\]

---

### 11. What are the characteristics of a normal distribution?

- Bell-shaped curve.
- Symmetrical about the mean.
- Mean = Median = Mode.
- Defined by mean \( \mu \) and standard deviation \( \sigma \).
- 68%-95%-99.7% rule applies.

---

### 12. What is the standard normal distribution, and why is it important?

A special case of normal distribution with:
- Mean = 0
- Standard deviation = 1

Used to standardize any normal distribution for easier probability calculations via Z-scores.

---

### 13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?

CLT states that the sampling distribution of the sample mean approaches a normal distribution as sample size increases, regardless of the population’s distribution.

It's important because it allows the use of normal distribution for inference.

---

### 14. How does the Central Limit Theorem relate to the normal distribution?

CLT explains why the **sample mean** is normally distributed even if the population is not, making it easier to apply Z-tests and confidence intervals.

---

### 15. What is the application of Z statistics in hypothesis testing?

Z-statistics helps compare sample results to population parameters under the assumption of a normal distribution. It's used in:
- One-sample and two-sample Z-tests.
- Confidence intervals.

---

### 16. How do you calculate a Z-score, and what does it represent?

Z-score:
\[
Z = \frac{X - \mu}{\sigma}
\]
It represents how many standard deviations a value \( X \) is from the mean \( \mu \).

---

### 17. What are point estimates and interval estimates in statistics?

- **Point Estimate**: A single value (like sample mean) used to estimate population parameter.
- **Interval Estimate**: A range (like confidence interval) within which the parameter lies.

---

### 18. What is the significance of confidence intervals in statistical analysis?

Confidence intervals show the reliability of an estimate. A 95% CI means that if we repeat the sampling process, 95% of intervals will contain the true population parameter.

---

### 19. What is the relationship between a Z-score and a confidence interval?

Z-scores define the width of confidence intervals. For example:
- 95% CI uses Z = 1.96
- 99% CI uses Z = 2.576

---

### 20. How are Z-scores used to compare different distributions?

Z-scores standardize different distributions, allowing comparisons by showing how extreme a value is in its own distribution.

---

### 21. What are the assumptions for applying the Central Limit Theorem?

- Random sampling.
- Independent samples.
- Sample size ≥ 30 for non-normal populations.
- Finite population variance.

---

### 22. What is the concept of expected value in a probability distribution?

Expected value is the theoretical average:
\[
E(X) = \sum x_i P(x_i) \text{ (discrete) or } E(X) = \int x f(x) dx \text{ (continuous)}
\]

---

### 23. How does a probability distribution relate to the expected outcome of a random variable?

The probability distribution gives the weights (probabilities) to outcomes. The expected value is the weighted average of all outcomes.
